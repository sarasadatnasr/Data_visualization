```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

# Activity 4: Generate a Bar Plot and a Heatmap to Represent Content Rating Types

```python
# Download dataset from the GitHub repository
apps_df = pd.read_csv('googleplaystore.csv')

# Display the first few rows of the DataFrame
print(apps_df.head())
```

```python
apps_df.dropna(inplace=True)
# Display the modified DataFrame
print(apps_df.head())
```

```python
# Count the number of apps in each Content Rating category
content_rating_counts = apps_df['Content Rating'].value_counts()

# Create the bar plot
plt.bar(content_rating_counts.index, content_rating_counts.values)

# Add labels and title
plt.xlabel('Content Rating Category')
plt.ylabel('Number of Apps')
plt.title('Number of Apps in Each Content Rating Category')

# Display the bar plot
plt.show()
```

```python
# Prepare the data for the heatmap
heatmap_data = apps_df.groupby(
    ['Category', 'Rating']).size().unstack().fillna(0)

# Create the heatmap
sns.heatmap(heatmap_data, cmap='YlGnBu')

# Add labels and title
plt.xlabel('Rating')
plt.ylabel('Category')
plt.title('Number of Apps by Category and Rating')

# Display the heatmap
plt.show()
```

```python
# Create a subplot with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the bar chart
ax1.bar(content_rating_counts.index, content_rating_counts.values)
ax1.set_xlabel('Content Rating Category')
ax1.set_ylabel('Number of Apps')
ax1.set_title('Number of Apps in Each Content Rating Category')

# Plot the heatmap
sns.heatmap(heatmap_data, cmap='YlGnBu', ax=ax2)
ax2.set_xlabel('Rating')
ax2.set_ylabel('Category')
ax2.set_title('Number of Apps by Category and Rating')

# Link the plots dynamically
plt.tight_layout()
plt.show()
```
