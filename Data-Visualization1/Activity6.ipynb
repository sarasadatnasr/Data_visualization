```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

# Activity 6: Creating a Choropleth Map to Represent Total Renewable Energy

```python
# Load renewable energy production dataset
production_data = pd.read_csv('share-electricity-renewables.csv')
print(production_data)
```

```python
# Sort DataFrame by the Year feature
production_data.sort_values('Year', inplace=True)
print(production_data.head())
```

```python
# Create the choropleth map
fig = px.choropleth(production_data,
                    locations='Code',
                    locationmode="ISO-3",
                    color='Renewables (% electricity)',
                    hover_name='Entity',
                    animation_frame='Year',
                    title='Renewable Energy Production')
fig.show()
```

```python
# Update projection style
fig.update_geos(projection_type="natural earth")

# Show the figure
fig.show()
```
