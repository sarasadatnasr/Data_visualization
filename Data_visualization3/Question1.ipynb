# Bar Plot

```python
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as py
```

```python
dataset1 = pd.read_csv("covid.csv")
dataset2 = pd.read_csv("covid_grouped.csv")
```

```python
country_cases = dataset1.groupby('Country/Region')['TotalCases'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
country_cases.plot(kind='bar')
plt.xlabel('Country/Region')
plt.ylabel('Total Cases')
plt.title('Total COVID-19 Cases by Country/Region')
plt.show()
```

```python
continent_deaths = dataset1.groupby('Continent')['TotalDeaths'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
continent_deaths.plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Total Deaths')
plt.title('Total COVID-19 Deaths by Continent')
plt.show()
```

```python
who_active_cases = dataset1.groupby('WHO Region')['ActiveCases'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
who_active_cases.plot(kind='bar')
plt.xlabel('WHO Region')
plt.ylabel('Active Cases')
plt.title('Active COVID-19 Cases by WHO Region')
plt.show()
```

```python

country_cases = dataset1.groupby('Country/Region')['TotalCases'].sum()
country_cases = country_cases.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.plot(country_cases.index, country_cases.values)
plt.xlabel('Country/Region')
plt.ylabel('Total Cases')
plt.title('Total COVID-19 Cases by Country/Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
