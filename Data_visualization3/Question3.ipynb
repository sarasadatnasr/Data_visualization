# Scatter and Line Plot

```python
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as py
```

```python
dataset2 = pd.read_csv("covid_grouped.csv")
dataset1 = pd.read_csv("covid.csv")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataset1['TotalCases'], dataset1['TotalDeaths'])
plt.scatter(dataset1['TotalCases'], dataset1['TotalDeaths'])
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Total COVID-19 Cases vs. Total Deaths')
plt.show()
```

```python
scat = px.scatter(dataset1, x='TotalCases', y='TotalDeaths', color='Country/Region', marginal_y='box')
scat.show()
```

```python
density_contour_plot = px.density_contour(dataset1, x="TotalCases", y="TotalDeaths", color="Country/Region",
                                          marginal_x="rug", marginal_y="histogram")

density_contour_plot.show()
```
