# Heatmap

```python
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as py
import seaborn as sns
```

```python
dataset1 = pd.read_csv("covid.csv")
```

```python
content_rating_counts = dataset1['Continent'].value_counts()
heatmap_data = dataset1.pivot_table(index='Country/Region', columns='Continent', values='TotalCases')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.bar(content_rating_counts.index, content_rating_counts.values)
ax1.set_xlabel('Continent')
ax1.set_ylabel('Number of Countries')
ax1.set_title('Number of Countries in Each Continent')

sns.heatmap(heatmap_data, cmap='YlGnBu', ax=ax2)
ax2.set_xlabel('Continent')
ax2.set_ylabel('Country/Region')
ax2.set_title('Total Cases by Country/Region and Continent')

plt.tight_layout()
plt.show()
```

```python
heatmap_data = dataset1.pivot_table(index='Country/Region', columns='Continent', values='TotalCases', fill_value=0)

sns.heatmap(heatmap_data, cmap='YlGnBu')

plt.xlabel('Continent')
plt.ylabel('Country/Region')
plt.title('Number of Total Cases by Country/Region and Continent')

plt.show()
```
